```python
import csv
import os
import requests

players_ids_src = os.path.join('..', '..', 'data', 'players', 'players_ids.csv')
players_rating = []
with open(players_ids_src, "r") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        player_id = row['id_sofascore']
        if player_id == "":
            player_ranking = {"sofascore_id": player_id, "events_rating": ""}
        events_response = requests.get(f'https://api.sofascore.com/api/v1/player/{player_id}/events/last/0',
                                       headers=headers).json()
        events_data = events_response.get("events", [])
        event_list = []
        for event in events_data:
            tournament_id = event.get("tournament", {}).get("UniqueTournament", {}).get("id", "")
            season_id = event.get("id", "")
            ratings_response = requests.get(
                f'https://api.sofascore.com/api/v1/player/{player_id}/unique-tournament/{tournament_id}/season/{season_id}/last-ratings',headers=headers).json()
            ratings = ratings_response.get("lastRatings", [])
            rating_list = []
            for rating in ratings:
                event_rating = rating.get("rating", "")
                rating_data = {'rating': event_rating}
                rating_list.append(rating_data)
            event_data = {'tournament_id': tournament_id, "season_id": season_id, "ratings": rating_list}
            event_list.append(event_data)
        player_rating = {"sofascore_id": player_id, "events_rating": event_list}
        players_rating.append(player_rating)
            
for index, obiekt in enumerate(players_rating[:10]):
    print(f"Obiekt {index + 1}: {obiekt}")
```
